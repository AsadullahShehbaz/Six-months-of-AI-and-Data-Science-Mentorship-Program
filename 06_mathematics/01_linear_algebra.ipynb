## 1.Matrix Equation AX=B

In linear algebra, the equation AX = B is called a matrix equation or a system of linear equations in matrix form

Here's what it represents:

- A is a matrix (often called the coefficient matrix),

- X is a column vector of variables (unknowns),

- B is a column vector of constants (also called the right-hand side vector).
  
System of Linear Equation : 

5x+2y=1<br>
3x+4y=2

```python
# linear algebra in python
import numpy as np
# create coefficient matrix 
A = np.array([[5, 2], [3, 4]])
print(f"A = {A}")
# create constant matrix B
B = np.array([1,2])
print(f"B = {B}")
# Solve for X
X = np.linalg.solve(A,B)
print("X = ", X)
```

## 2.Complex Equations

**(1+2i) x+(2-i) y = 4+i**<br>
**(4+i) x+(3+4i) y = 3+4i**

```python
# Coefficients matrix A with complex numbers 
A_complex = np.array([[1+2j, 2-1j], [3-1j, 4]]) 
# Constants matrix B with complex numbers 
B_complex = np.array([4+1j, 3+4j]) 
# Solving for X in the complex case 
X_complex = np.linalg.solve (A_complex, B_complex) 
X_complex
```

## 3.Vectors in Python

```python
import numpy as np
import matplotlib.pyplot as plt

# create a vector 
vector = np.array([2,3])
print(f"Vector : {vector}")

# Plot the vector 
plt.figure()
plt.quiver(0,0,vector[0],vector[1],angles='xy',scale_units='xy',scale=1,color='r')
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()
plt.title('2D Vector')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

## 4.Linear Transformation in Python 

\[
\begin{bmatrix}
\cos(45^\circ) & \sin(45^\circ)\\
\sin(45^\circ) & \cos(45^\circ)
\end{bmatrix}
\]

```python
#Define the rotation matrix for 45 degree rotation 
theta = np.radians (45) # Converting 45 degrees to radians 
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)], 
[np.sin(theta), np.cos(theta)]]) 
#Applying the transformation 
v_transformed = np.dot(rotation_matrix, vector) 
# Plotting the original and transformed vectors 
plt.figure() 
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector') 
plt.quiver(0, 0, v_transformed [0], v_transformed [1], angles='xy', scale_units='xy', scale=1, color='b', label='Transform Vector')
plt.xlim(-1, 5) 
plt.ylim(-1, 5) 
plt.grid() 
plt.title('20 Vector Transformation') 
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
