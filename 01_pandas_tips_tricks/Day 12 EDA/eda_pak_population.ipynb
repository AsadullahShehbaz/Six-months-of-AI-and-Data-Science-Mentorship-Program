```python
# total rural-population of pakistan
print(f"Total rural population in Pakistan(2017) is {df['ALL SEXES (RURAL)'].sum()}")
```

```python
# total rural-population of pakistan
print(f"Total rural population in Pakistan(1998) is {df['POPULATION 1998 (RURAL)'].sum()}")
```

```python
# Difference in rural population of Pakistan from 1998 to 2017
rural_pop_1998 = df['POPULATION 1998 (RURAL)'].sum()
rural_pop_2017 = df['ALL SEXES (RURAL)'].sum()
pop_change = rural_pop_2017 - rural_pop_1998
print(f"Total rural population increased from 1998-2017 : {pop_change}")
print(f"Total rural population increased from 1998-2017 : {pop_change/rural_pop_1998*100:.2f}%")
```

```python
# Difference in rural population of Pakistan from 1998 to 2017
URBAN_pop_1998 = df['POPULATION 1998 (URBAN)'].sum()
URBAN_pop_2017 = df['ALL SEXES (URBAN)'].sum()
pop_change = URBAN_pop_2017 - URBAN_pop_1998
print(f"Total URBAN population increased from 1998-2017 : {pop_change}")
print(f"Total URBAN population increased from 1998-2017 : {pop_change/URBAN_pop_1998*100:.2f}%")
```
