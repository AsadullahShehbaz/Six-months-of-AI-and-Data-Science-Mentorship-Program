```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
```

# 1.Load Data

```python
df = sns.load_dataset('titanic')
```

# 2.Select the X Axis and Y Axis

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='age', y='fare', data=df,hue='sex')
plt.xlablel('Age')
plt.ylable('Fare')
plt.title('Age vs Fare')
```

# 3.Use Manual Colors

```python
plt.figure(figsize=(16,9))
palette={'male':'orange','female':'green'}
sns.scatterplot(x='age', y='fare', data=df,hue='sex',palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare')
```

# 4.Styling Plot

```python
sns.set_style('darkgrid')
sns.scatterplot(x='age', y='fare', data=df,hue='sex')
```

```python
sns.set_theme(style='darkgrid', font_scale=1.5)
sns.scatterplot(x='age', y='fare', data=df,hue='sex')
```

## Scatter Plot with Plotly 

```python
px.scatter(df,x='age', y='fare',color='sex')
```

Interpretation
1. We have age of travellers at X-axis and fare at Y-axis
2. The scatter plot shows the relationship between age and fare of titanic passengers
3. We further grouped the passengers by their genders such as blue points for male and red points for female
4. Typo male passengers gave the highest fare in titanic and they might be travelling first class
5. The oldest person in the titanic passengers is also male having 80 years age as shown in the scatter plot 

## Bubble Plot 
- Add Size Parameter in the scatter plot to make bubble plot 

```python
plt.figure(figsize=(16,9))
sns.set_theme(style='darkgrid', font_scale=1)
sns.scatterplot(x='age', y='fare', data=df,hue='sex',size='pclass')
```

---
## Make Box Plot of Sex and Age and group by class

```python
plt.figure(figsize=(12,8))
sns.set_style('darkgrid')
sns.boxplot(data=df,x='sex',y='age',hue='class')
plt.legend(loc='upper right',title ='Class Name')
```

---
## Boxen Plot 

```python
plt.figure(figsize=(12,8))
sns.set_style('darkgrid')
sns.boxenplot(data=df,x='sex',y='age',hue='class')
plt.legend(loc='upper right',title ='Class Name')
plt.xticks(ticks=[0,1],labels=['Female','Male'])
plt.xticks(rotation=45)
```

## Bar Plot with Capsize

```python
sns.barplot(data=df,x='sex',y='age',hue='class',capsize=0.3,ci=None)
```

**Interpretation:**

This creates a bar plot showing average fare per sex, with:
1. no confidence intervals (**ci=None**)
2. short caps on the (nonexistent) error bars (**capsize=0.3**  has no effect here since **ci=None**).
